## Embeddings

#### Embedding Techniques

Embeddings are a numerical representation of text that can be used to measure the relatedness between two pieces of text. Embeddings are useful for search, clustering, recommendations, anomaly detection, and classification tasks.

````python
import os
from dotenv import load_dotenv
load_dotenv()  #load all the environment variables
````

````python
from langchain_openai import OpenAIEmbeddings
embeddings=OpenAIEmbeddings(model="text-embedding-3-small")
embeddings
````

````python
text="This is a tutorial on OPENAI embedding"
query_result=embeddings.embed_query(text)
query_result
````

````python
len(query_result)
````

````python
# You can also embed a list of documents
documents = ["This is a document about embeddings.", "Another document for embedding.","cat"]
document_embeddings = embeddings.embed_documents(documents)
````

````python
print(f"\nEmbeddings for the documents:")
for i, emb in enumerate(document_embeddings):
    print(f"Document {i+1} embedding length: {len(emb)}")
````

````python
embeddings_1024=OpenAIEmbeddings(model="text-embedding-3-large",dimensions=1024)
embeddings_1024
````

````python
text="This is a tutorial on OPENAI embedding"
query_result=embeddings_1024.embed_query(text)
len(query_result)
````

````python
from langchain_huggingface import HuggingFaceEmbeddings
````

````python
# Instantiate the HuggingFaceEmbeddings model
hf_embeddings = HuggingFaceEmbeddings(model_name="all-MiniLM-L6-v2")
````

````python

query = "This is a tutorial on OPENAI embedding"

# Use the correct method to embed a single query string
query_embedding = hf_embeddings.embed_query(query)

print(f"\nEmbedding for the query: {query}")
print(f"Embedding length: {len(query_embedding)}")
````
