```python
from langchain_openai import ChatOpenAI
from dotenv import load_dotenv
from langchain_core.prompts import PromptTemplate
from langchain_core.output_parsers import StrOutputParser

load_dotenv()
```

```python
## SimpleLLM + Prompt + Parser chain using LCEL
prompt = PromptTemplate(
    template='write a joke about the {topic}',
    input_variables=['topic']
)

model = ChatOpenAI()

parser = StrOutputParser()

# Create the combined chain using LangChain Expression Language (LCEL)
chain = prompt | model | parser

result = chain.invoke({'topic':'python'})

print(result)

chain.get_graph().print_ascii()
```

## Runnable Parallel

```python
from langchain_openai import ChatOpenAI
from langchain_core.prompts import PromptTemplate
from langchain_core.output_parsers import StrOutputParser
from dotenv import load_dotenv
from langchain_classic.schema.runnable import RunnableSequence, RunnableParallel

load_dotenv()
```

```python
prompt1 = PromptTemplate(
    template='Generate a tweet about {topic}',
    input_variables=['topic']
)
```

```python
prompt2 = PromptTemplate(
    template='Generate a Linkedin post about {topic}',
    input_variables=['topic']
)
```

```python
model = ChatOpenAI()
parser = StrOutputParser()
```

```python
parallel_chain = RunnableParallel(
    {
        'tweet': RunnableSequence(prompt1, model,parser),
        'linkedin_post': RunnableSequence(prompt2, model,parser),
    }
)

result = parallel_chain.invoke({'topic': 'AI'})
```

```python
print(f"Tweet: {result['tweet']}")
print("*"*50)
print(f"Linkedin Post: {result['linkedin_post']}")
```

## Runnable Lambda

Converting any function to a runnable using RunnableLambda

```python
from langchain_openai import ChatOpenAI
from langchain_core.prompts import PromptTemplate
from langchain_core.output_parsers import StrOutputParser
from dotenv import load_dotenv
from langchain_classic.schema.runnable import RunnableSequence, RunnableLambda, RunnablePassthrough, RunnableParallel

load_dotenv()
```

```python
def word_count(text):
    return len(text.split())

# convert function function to Runnable

runnable_word_count = RunnableLambda(word_count)
print(runnable_word_count.invoke("Hello World"))
```

```python
runnable_word_count = RunnableLambda(lambda x: len(x.split()))
print(runnable_word_count.invoke("Hello World"))
```

## Runnable Lambda Part2

```python
def word_count(text):
    return len(text.split())
```

```python
prompt = PromptTemplate(
    template='Write a joke about {topic}',
    input_variables=['topic']
)
```

```python
model = ChatOpenAI()
parser = StrOutputParser()
```

## Runnable Branch

```python
from langchain_openai import ChatOpenAI
from langchain_core.prompts import PromptTemplate
from langchain_core.output_parsers import StrOutputParser
from dotenv import load_dotenv
from langchain_classic.schema.runnable import RunnableSequence, RunnableParallel, RunnablePassthrough, RunnableBranch, RunnableLambda

load_dotenv()
```

```python
prompt1 = PromptTemplate(
    template='Write a detailed report on {topic}',
    input_variables=['topic']
)
```

```python
prompt2 = PromptTemplate(
    template='Summarize the following text \n {text}',
    input_variables=['text']
)
```

```python
model = ChatOpenAI()

parser = StrOutputParser()
```

```python
report_gen_chain = RunnableSequence(prompt1,model,parser)
result = report_gen_chain.invoke({'topic':'Evolution of Generative AI technologies and their impact on various industries over the past decade'})
print(result)
```

```python
branch_chain = RunnableBranch(
    (lambda x: len(x.split())>300, RunnableSequence(prompt2,model,parser)),
     RunnablePassthrough()     
)
```

```python
final_chain = RunnableSequence(report_gen_chain, branch_chain)

result = final_chain.invoke({'topic':'Evolution of Generative AI technologies and their impact on various industries over the past decade'})
print(result)
```
